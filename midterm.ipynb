#### 2.3. Figure out on what dates each stock had the best and worst single-day returs on returns data

```python
aapl_best = aapl_df.iloc[aapl_df['Return'].idxmax()]
aapl_worst = aapl_df.iloc[aapl_df['Return'].idxmin()]

nvda_best = nvda_df.iloc[nvda_df['Return'].idxmax()]
nvda_worst = nvda_df.iloc[nvda_df['Return'].idxmin()]

tgt_best = tgt_df.iloc[tgt_df['Return'].idxmax()]
tgt_worst = tgt_df.iloc[tgt_df['Return'].idxmin()]

tsla_best = tsla_df.iloc[tsla_df['Return'].idxmax()]
tsla_worst = tsla_df.iloc[tsla_df['Return'].idxmin()]

amzn_best = amzn_df.iloc[amzn_df['Return'].idxmax()]
amzn_worst = amzn_df.iloc[amzn_df['Return'].idxmin()]
```

```python
task23 = pd.DataFrame({
    'Stock': ['AAPL', 'NVDA', 'TGT', 'TSLA', 'AMZN'],
    'Best Date': [aapl_best['Date'].date(), 
                  nvda_best['Date'].date(), 
                  tgt_best['Date'].date(), 
                  tsla_best['Date'].date(), 
                  amzn_best['Date'].date()],
    'Best Return': [aapl_best['Return'], 
                    nvda_best['Return'], 
                    tgt_best['Return'], 
                    tsla_best['Return'], 
                    amzn_best['Return']],
    'Worst Date': [aapl_worst['Date'].date(), 
                   nvda_worst['Date'].date(), 
                   tgt_worst['Date'].date(), 
                   tsla_worst['Date'].date(), 
                   amzn_worst['Date'].date()],
    'Worst Return': [aapl_worst['Return'], 
                     nvda_worst['Return'], 
                     tgt_worst['Return'], 
                     tsla_worst['Return'], 
                     amzn_worst['Return']]
})

display(task23)
```

#### 2.4. Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2020?

```python
std24 = pd.DataFrame({
    'Stock': ['AAPL', 'NVDA', 'TGT', 'TSLA', 'AMZN'],
    'STD Full': [aapl_df['Return'].std(), 
                 nvda_df['Return'].std(), 
                 tgt_df['Return'].std(), 
                 tsla_df['Return'].std(), 
                 amzn_df['Return'].std()],
    'STD 2020': [aapl_df[aapl_df['Date'].dt.year == 2020]['Return'].std(), 
                 nvda_df[nvda_df['Date'].dt.year == 2020]['Return'].std(), 
                 tgt_df[tgt_df['Date'].dt.year == 2020]['Return'].std(), 
                 tsla_df[tsla_df['Date'].dt.year == 2020]['Return'].std(), 
                 amzn_df[amzn_df['Date'].dt.year == 2020]['Return'].std()]
})

display(std24)
```

```python
max_std_full_stock = std24.loc[std24['STD Full'].idxmax()]
print(max_std_full_stock['Stock'], "has max STD of all time with STD =", max_std_full_stock['STD Full'])
```

```python
max_std_2020_stock = std24.loc[std24['STD 2020'].idxmax()]
print(max_std_full_stock['Stock'], "has max STD of 2020 with STD =:", max_std_full_stock['STD Full'])
```

Based on the result, we can classify that $\textbf{TSLA}$ is $\textbf{the riskiest}$ stock for both $\textbf{entire time period}$ and in $\textbf{2020}$ categories since the standard deviation for both categories is the highest compares to the other four, which means the price fluctuated a lot.
