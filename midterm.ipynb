### **4. Data Insights**: Propose methods to predict the `Open` prices at a point in time in the future.

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np
```

```python
# drop NaN
meta_df = meta_df.dropna()
meta_df.head()
```

```python
features = ['Close', 'High', 'Low', 'Volume', 'Adj Close', 'Return']
target = 'Open'
```

```python
meta_df.set_index('Date', inplace=True)
meta_df.head()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(meta_df[features + [target]])
```

```python
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length, :-1])
        y.append(data[i + seq_length, -1])
    return np.array(X), np.array(y)

seq_length = 60
X, y = create_sequences(scaled_data, seq_length)
```

```python
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(seq_length, X_train.shape[2])))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)
```

```python
loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')
```

```python
predictions = model.predict(X_test)
predictions_rescaled = scaler.inverse_transform(
    np.concatenate((np.zeros((predictions.shape[0], scaled_data.shape[1] - 1)), predictions), axis=1))[:, -1]
```

```python
y_test_rescaled = scaler.inverse_transform(
    np.concatenate((np.zeros((y_test.shape[0], scaled_data.shape[1] - 1)), y_test.reshape(-1, 1)), axis=1)
)[:, -1]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(meta_df.index[-len(y_test_rescaled):], y_test_rescaled, label='Actual Open Price')
plt.plot(meta_df.index[-len(predictions_rescaled):], predictions_rescaled, label='Predicted Open Price')
plt.title('Actual vs Predicted Open Prices')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.legend()
plt.show();
```

To predict the open prices of stocks at a future point in time, various methods can be employed, with $\textbf{Long Short-Term Memory (LSTM)}$ networks being one of the most effective approaches for time-series forecasting. LSTM models are well-suited for this task due to their ability to handle sequential data and capture long-term dependencies in the stock price movements. By learning from historical price patterns and trends, LSTMs can predict future open prices based on the information from past prices, market behavior, and other factors. Unlike traditional models that treat data points independently, LSTMs retain memory of previous time steps, making them particularly adept at identifying non-linear relationships and filtering out noise. Therefore, leveraging LSTMs allows for more accurate and robust predictions of stock open prices, especially when the data exhibits complex patterns and dependencies over time.

```python
def predict_future_prices(model, data, scaler, seq_length, num_predictions):
    last_sequence = data[-seq_length:, :-1]
    predicted_prices = []

    for _ in range(num_predictions):
        last_sequence_reshaped = last_sequence.reshape((1, seq_length, data.shape[1] - 1))
        next_pred = model.predict(last_sequence_reshaped)[0]
        predicted_prices.append(next_pred[0])
        next_sequence = np.append(last_sequence[1:], [[*last_sequence[-1, :-1], next_pred[0]]], axis=0)
        last_sequence = next_sequence

    return np.array(predicted_prices)
```

```python
num_future_days = 20
future_predictions = predict_future_prices(model, scaled_data, scaler, seq_length, num_future_days)
```

```python
future_predictions_rescaled = scaler.inverse_transform(
    np.concatenate((np.zeros((future_predictions.shape[0], scaled_data.shape[1] - 1)), future_predictions.reshape(-1, 1)), axis=1)
)[:, -1]
```

```python
last_date = pd.to_datetime(meta_df.index[-1])
future_dates = [last_date + pd.Timedelta(days=i) for i in range(1, num_future_days + 1)]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(meta_df.index[-len(y_test_rescaled):], y_test_rescaled, label='Actual Open Price')
plt.plot(meta_df.index[-len(predictions_rescaled):], predictions_rescaled, label='Predicted Open Price')
plt.plot(future_dates, future_predictions_rescaled, label='Future Predicted Open Price')
plt.title('Actual, Predicted, and Future Predicted Open Prices')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.legend()
plt.show();
```
