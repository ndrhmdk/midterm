## **2.2. Create a new DataFrame, titled "$\textbf{Return}$". This DataFrame will contain the returns for each stock data that you had crawled above. Returns are typically defined by**
$$r_t=\frac{p_t - p_{t-1}}{p_{t-1}}=\frac{p_t}{p_{t-1}-1}$$

#### **Apple Inc. (AAPL)**

## **2.3. Figure out on what dates each stock had the best and worst single-day returns on returns data**

```python
stocks = ['Apple Inc. (AAPL)', 'NVDIA Corporation (NVDA)', 'Ford Motor Corporation (F)',
          'Microsoft Corporation (MSFT)', 'Meta Platform, Inc. (META)']
dfs = [aapl, nvda, ford, msft, meta]

data = {'Stock': [], 
        'Best Date': [], 'Best Return': [],
        'Worst Date': [], 'Worst Return': []}

for stock, df in zip(stocks, dfs):
    best = df.iloc[df['Return'].idxmax()]
    worst = df.iloc[df['Return'].idxmin()]
    
    data['Stock'].append(stock)
    data['Best Date'].append(best['Date'].date())
    data['Best Return'].append(best['Return'])
    data['Worst Date'].append(worst['Date'].date())
    data['Worst Return'].append(worst['Return'])
```

```python
display(pd.DataFrame(data))
```

## **2.4. Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2020?**

```python
data = {'Stock': stocks, 'STD (overall)': [], 'STD (2020)': []}
for df in dfs:
    data['STD (overall)'].append(df['Return'].std())
    data['STD (2020)'].append(df[df['Date'].dt.year == 2020]['Return'].std())
    
display(pd.DataFrame(data))
```

The highest value of standard deviation of the `Return` in both categories $\textit{\textbf{entire time period}}$ and $\textit{\textbf{in 2020}}$ is $\textbf{NVDIA Corporation (NVDA)}$. 

Standard deviation measures the volatility or dispersion of a stock's returns from its average return. A higher standard deviation indicates that the stock's returns fluctuate more, meaning it has greater unpredictability and risk. Stocks with high standard deviation are more volatile, meaning their price swings are larger and less predictatble. 

Therefore, $\textbf{NVDIA Corporation (NVDA)}$ would be classify as the riskiest stock over the entire time period and for the year 2020.

# **4. Data Insights**
## **4.1. Propose methods to predict the Open prices at a point in time in the future.**

```python
msft = yf.download('MSFT', start=starttime, end=endtime)
msft.columns = msft.columns.droplevel('Ticker')
msft.head()
```

```python
plt.figure(figsize=(16, 6))
plt.title('Close Price History')
plt.plot(msft['Close'], color='#E4C087')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

```python
data = msft.filter(['Close'])
dataset = data.values
training_data_len = int(np.ceil(len(dataset) * 0.95))
training_data_len
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
scaled_data
```

```python
train_data = scaled_data[0:int(training_data_len), :]
```

```python
X_train = []
y_train = []

for i in range(60, len(train_data)):
    X_train.append(train_data[i - 60: i, 0])
    y_train.append(train_data[i, 0])
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train.shape, y_train.shape
```

We use $\textbf{LSTM (Long Short-Term Memory)}$ for predicting the `Open` price in the future because $\textbf{LSTM}$ can capture complex patterns and dependencies in historical stock data, enabling it to make predictions about future price movements.

```python
from tensorflow import keras

model = keras.Sequential([
    keras.layers.LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    keras.layers.LSTM(64, return_sequences=False),
    keras.layers.Dense(25),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(X_train, y_train, batch_size=1, epochs=1)
```

```python
test_data = scaled_data[training_data_len - 60: , :]
```

```python
X_test = []
y_test = dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    X_test.append(test_data[i - 60: i, 0])

X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
```

```python
rmse = round(np.sqrt(np.mean(((predictions - y_test) ** 2))), 3)
print("Root Mean Squared Error:", rmse)
```

```python
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
```

```python
import matplotlib.dates as mdates
plt.figure(figsize=(16, 6))

plt.title('Prediction')
plt.xlabel('Date')
plt.ylabel('Close Price')

plt.plot(train['Close'], color='#E4C087', label='Train')
plt.plot(valid['Close'], color='#659287', label='Val')
plt.plot(valid['Predictions'], color='#AE445A', label='Predictions')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.xticks(rotation=45)

plt.legend(['Train', 'Val', 'Predictions'])
plt.grid(True)
plt.show();
```
